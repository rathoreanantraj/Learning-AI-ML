## 1. Output
The first and the easiest thing that we can probably do in Python is printing (or outputting) something on our screen. We can perform this simple task using a special function, called the `print()` function. 

* Think of a function as a specialized helper that performs a specific task.
* `print()` helps us display information, results of calculations, or formatted text on the console.
* Anything in single quotes (`'`) or double-quotes (`"`) is called a string and gets printed as it is.

```python
# Printing Strings
print("hello world")

# Printing numerical values
print(3.14)
print(2 + 3)

# Displaying multiple values on the same line
print("hello world", "from Day 2")
```

## 2. Variables
A variable is like a container that stores data or values in our program. We can think of it as a labeled box where we can place something in, change it, or use it later.

**Important Rules & Notes:**
1. **Dynamic Typing:** Python is dynamically typed. When creating variables, we don't need to declare their type explicitly.
2. **Case-Sensitive:** Variables like `age` and `Age` are treated differently.
3. **Naming Rules:** Names must start with a letter (a-z, A-Z) or underscore and cannot be Python keywords (like `if`, `for`, `class`).

```python
name = "Anant"
age = 30
PI = 3.14
word = "artificial intelligence"

print("value of PI is:", PI)
```

## 3. Data Types
A data type defines the kind of value a variable can hold. 

The fundamental built-in types include:
* `int`: integer values (whole numbers).
* `float`: floating-point values (decimal numbers).
* `str`: strings (sequence of characters like words & sentences).
* `bool`: boolean values (True / False).
* `None`: special constant representing no value / null.

You can check a variable's type using the `type()` function.

```python
x = 10
print(type(x)) # <class 'int'>

PI = 3.14
print(type(PI)) # <class 'float'>

name = "Anant"
print(type(name)) # <class 'str'>

isLearning = True
print(type(isLearning)) # <class 'bool'>

empty_var = None
print(type(empty_var)) # <class 'NoneType'>
```

## 4. Type Conversion & Type Casting
Type conversion is when we convert(cast) variables from one type to another. 

1. **Implicit (Type Conversion):** Done automatically by Python to prevent data loss.
2. **Explicit (Type Casting):** Done manually by the programmer using functions like `int()`, `float()`, `str()`, `bool()`.

```python
# 1. Implicit Conversion
a = 5
b = 3.0
print(a + b) # Python converts the answer into float by default

# 2. Explicit Type Casting
x = 10       # Current type: int
y = float(x) # Convert to float
z = str(x)   # Convert to string

print(y) 
print(z) 
print(type(z)) # This will output as <class 'str'>
```

## 5. Operators
An operator is a symbol that tells Python to perform a specific computation on one or more values (called operands).

1. **Arithmetic:** `+`, `-`, `*`, `/`, `%` (modulo/remainder), `**` (power).
2. **Relational:** `==`, `!=`, `>`, `<`, `>=`, `<=` (used to compare values).
3. **Assignment:** `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`.
4. **Logical:** `and`, `or`, `not`.

```python
a = 5
b = 10

# Arithmetic
print("Addition:", a + b)
print("Modulo (Remainder):", b % a)
print("Power:", a ** 2)

# Relational
print("Is a equal to b?:", a == b)
print("Is a not equal to b?:", a != b)

# Logical
x = 10
y = 20
z = 5
print("AND operator:", x > z and y > x) # True because BOTH conditions are true
```

## 6. Input & Practice Problem
We use the `input()` function to take input from the user while the program is running.
**Crucial Note:** Whatever the user types in the input prompt is returned as a string by default. We must cast it if we want to do math with it!

**Practice Problem:** Write a program that takes input of 2 numbers (`a` & `b`) and print the average.

```python
# The input is cast directly to an integer using int()
a = int(input("enter 1st num: ")) 
b = int(input("enter 2nd num: ")) 

# Utilizing parentheses to enforce operator precedence
avg = (a + b) / 2 

print("average =", avg)
```
