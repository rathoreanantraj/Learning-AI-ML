## 1. Conditional Statements (Decision Making)
Conditional statements allow our program to decide what action to take based on specific conditions (True or False expressions). 

There are 3 main building blocks for conditionals in Python:
1. `if`: Used to test a primary condition. If this expression evaluates to `True`, the indented block of code directly below it runs.
2. `elif` (Else If): Used to check additional conditions if the previous `if` or `elif` conditions were `False`. You can have as many `elif` statements as you want.
3. `else`: The fallback option. This block runs only if **all** preceding `if` and `elif` conditions evaluate to `False`.

*Note: `elif` and `else` cannot exist on their own. They must be preceded by an `if` statement.*

```python
# Example 1: Basic If-Else
print("--- Example 1: Voting Eligibility ---")
age = 20 # You can change this to test different outputs

if age >= 18:
    print("You can vote")
    print("You can drive")
else:
    print("You can't vote")
    print("You can't drive")

# Example 2: If-Elif-Else Chain
print("\n--- Example 2: Traffic Lights ---")
color = "yellow"

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Look")
elif color == "green":
    print("Go")
else:
    print("Wrong color entered")
```

## 2. Advanced Conditionals: Logical Operators & Nesting

### Logical Operators
Sometimes, a single condition isn't enough. We can combine multiple conditions using logical operators:
* `and`: Returns True only if **both** conditions are True.
* `or`: Returns True if **at least one** condition is True.
* `not`: Reverses the result (True becomes False, and vice versa).

### Nesting
Nesting means placing one block of code inside another. In conditionals, this means placing an `if` statement inside another `if` or `else` block. This is highly useful for multi-step decision-making, like checking a username first, and then checking a password.

```python
# Logical Operators in Conditionals
print("--- Age Categorization ---")
person_age = 15

if person_age < 13:
    print("Category: Child")
elif (person_age >= 13 and person_age < 18):
    print("Category: Teenager")
else:
    print("Category: Adult")

# Nesting Example: Basic Login System
print("\n--- Login System ---")
username = "admin"
password = "wrong_password"

if username == "admin" and password == "pass":
    print("Log in successful!")
else:
    # NESTING: We enter this else block, and then make another decision
    if username != "admin":
        print("Wrong user name, try again.")
    else:
        print("Wrong password, try again.")
```

## 3. Shorthand Conditionals: Ternary & Match Case

### Ternary Statements
Ternary statements provide a compact way to write simple `if-else` blocks in a single line of code. 
**Syntax:** `value_if_true if condition else value_if_false`
*Best Practice:* Only use this for very simple conditions to maintain readability.

### Match Case Statements (Introduced in Python 3.10)
If you have a long chain of `elif` statements checking the exact value of a single variable, `match` and `case` provide a cleaner alternative.
* `match`: Defines the variable you are checking.
* `case`: Defines the pattern/value to check against.
* `case _`: Acts as a wildcard/default case (exactly like `else`).

```python
# Ternary Statement Example
current_age = 18
status = "Adult" if current_age >= 18 else "Not Adult"
print("Ternary output:", status)

# Match Case Example
signal_color = "green"

match signal_color:
    case "red":
        print("Action: Stop")
    case "yellow":
        print("Action: Look")
    case "green":
        print("Action: Go")
    case _:
        print("Action: Invalid signal color")
```

## 5. Loops: The `while` Loop
A loop allows us to execute a block of code multiple times. Python has two main types of loops: `while` and `for`.

### The `while` Loop
The `while` loop repeats a block of code **as long as a specific condition remains True**. 
1. The condition is evaluated.
2. If True, the code block runs.
3. The condition is checked again, and the cycle repeats until the condition becomes False.

**Warning - Infinite Loops:** If your condition never becomes False (e.g., you forget to increase your counter variable), the loop will run forever and crash your program.

```python
# Example 1: Print numbers 1 to 5
print("--- Counting Up ---")
i = 1          # Step 1: Initialize a counter
while i <= 5:  # Step 2: Set the condition
    print(i)
    i += 1     # Step 3: Update the counter (Crucial to avoid infinite loops)

# Example 2: Print numbers 5 to 1
print("\n--- Counting Down ---")
i = 5
while i > 0:
    print(i)
    i -= 1     # Decrementing the counter
```

## 6. Loop Control Statements (`break` & `continue`)
Sometimes we need manual control over how a loop operates inside its code block.

* **`break`**: Immediately terminates and exits the entire loop. No further iterations will occur.
* **`continue`**: Immediately stops the *current* iteration, skips any remaining code below it in the loop body, and jumps straight to the next iteration.

```python
# Break Example: Stop loop when a multiple of 6 is found
print("--- Break Example ---")
i = 1
while i <= 10:
    if i % 6 == 0:
        print("Multiple of 6 found! Breaking the loop.")
        break  # Loop ends here completely
    print("Number:", i)
    i += 1

# Continue Example: Skip multiples of 3
print("\n--- Continue Example ---")
j = 0
while j < 10:
    j += 1
    if j % 3 == 0:
        continue # Skips the print statement for 3, 6, and 9
    print("Number:", j)
```

## 7. Loops: The `for` Loop & The `in` Keyword
Unlike the `while` loop (which relies on a boolean condition), the `for` loop is designed to **iterate over a sequence** (like a string of characters, a list, or a range of numbers).

### The `in` Keyword (Membership Operator)
The `in` keyword checks if an item exists within a sequence. In a `for` loop, it extracts items from the sequence one by one.

```python
word = "Prime"

# Example 1: Looping over a string
print("--- String Iteration ---")
for ch in word:
    print("Character:", ch)

# Example 2: Using 'in' outside a loop for simple checks
if 'i' in word:
    print("\nCheck: The letter 'i' exists in the word!")

# Example 3: Counting specific characters
search_word = "artificial intelligence"
count = 0
for ch in search_word:
    if ch == 'i':
        count += 1
print(f"\nThe letter 'i' occurs {count} times in '{search_word}'.")
```

## 8. The `range()` Function & Nested Loops

### The `range()` Function
`range()` generates a sequence of numbers. It is highly efficient because it creates an iterator (generating numbers on the fly) rather than saving a massive list in memory.
It takes up to 3 arguments: `range(start, stop, step)`
* **start:** (Optional, defaults to 0) The starting number.
* **stop:** (Required) The number to stop *before* (it is not included in the output).
* **step:** (Optional, defaults to 1) The increment value.

### Nested Loops
Just like nested `if` statements, you can put a loop inside another loop. The inner loop will finish all of its iterations for *every single iteration* of the outer loop.

```python
# Range with 1 argument (stop) - starts at 0
print("--- Range (Stop: 5) ---")
for i in range(5):
    print(i, end=" ") # end=" " prints on the same line

# Range with 2 arguments (start, stop)
print("\n\n--- Range (Start: 1, Stop: 6) ---")
for i in range(1, 6):
    print(i, end=" ")

# Range with 3 arguments (start, stop, step)
print("\n\n--- Range (Start: 1, Stop: 10, Step: 2) ---")
for i in range(1, 10, 2):
    print(i, end=" ")

# Nested Loops Example (Coordinates)
print("\n\n--- Nested Loops ---")
for i in range(1, 3):       # Outer loop runs 2 times (1, 2)
    for j in range(1, 3):   # Inner loop runs 2 times for each outer loop iteration
        print(f"Coordinate: ({i}, {j})")
```

## 9. Loops: Practice Problems (Set 2)
1. Print the multiplication table for a number `n` using `while`.
2. Print odd numbers from 1 to 10 using `continue` and `while`.
3. Count vowels in a given word using `for`.
4. Calculate the sum of the first `n` natural numbers using `for`.

```python
# Solution 1: Multiplication Table
print("--- Q1: Multiplication Table ---")
n = 5
i = 1
while i <= 10:
    print(f"{n} x {i} = {i * n}")
    i += 1

# Solution 2: Odd numbers using continue
print("\n--- Q2: Odd Numbers ---")
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i, end=" ")

# Solution 3: Count Vowels
print("\n\n--- Q3: Count Vowels ---")
test_word = "education"
vowels_count = 0
for ch in test_word:
    if ch in 'aeiouAEIOU': # Cleaner way to check multiple conditions
        vowels_count += 1
print(f"Vowels found: {vowels_count}")

# Solution 4: Sum of natural numbers
print("\n--- Q4: Sum of first N numbers ---")
limit = 10
total_sum = 0
for i in range(1, limit + 1): # limit + 1 ensures the limit itself is included
    total_sum += i
print(f"Sum of first {limit} numbers is: {total_sum}")
```

## 10. Functions: Making Code Reusable
Functions are blocks of code that perform a specific task. We use them so we don't have to rewrite the same logic over and over.

* **Defining a function:** Use the `def` keyword, followed by the function name and parentheses.
* **Calling a function:** Write the function's name followed by parentheses to execute it.
* **Parameters & Arguments:** Parameters act as variables inside the function parentheses. When you call the function, the actual data you pass into it are called Arguments.
* **The `return` keyword:** A function can optionally send processed data back to the rest of your program using `return`. Once `return` is executed, the function immediately ends.

```python
# Basic Function Definition
def greet():
    print("Hello from the greet function!")

greet() # Function Call

# Function with Parameters
def add_numbers(a, b): # 'a' and 'b' are parameters
    print(f"The sum is: {a + b}")

add_numbers(10, 20) # 10 and 20 are arguments

# Function returning a value
def compute_average(a, b, c):
    return (a + b + c) / 3

# We store the returned value in a variable
result = compute_average(10, 20, 30)
print(f"The computed average is: {result}")
```

## 11. Default Parameters & Built-in Functions

### Default Parameters
You can assign default values to parameters. If the person calling the function doesn't provide an argument for that parameter, Python will automatically use the default.
**Rule:** Default parameters must *always* come at the end of the parameter list. You cannot place a non-default parameter after a default one.

### Built-in Functions
Python provides many functions out of the box. We have already been using them! The underlying logic is handled by Python, we just call them.
Examples: `print()`, `input()`, `type()`, `range()`

```python
# Default Parameter Example
def greet_user(name, greeting="Hello"): 
    # 'greeting' has a default value. 'name' does not, so it must be provided.
    print(f"{greeting}, {name}!")

greet_user("Anant")             # Uses default greeting
greet_user("Anant", "Welcome")  # Overrides default greeting
```

## 12. Lambda Functions
Lambda functions are special, anonymous (nameless) functions used for writing short, one-liner logic. 

**Syntax:** `lambda arguments: expression`
* They can take any number of arguments.
* They can only contain a single expression.
* The result of the expression is automatically returned (no need to write the `return` keyword).

```python
# Standard function for squaring a number
def square_standard(x):
    return x * x

# Equivalent Lambda Function
square_lambda = lambda x: x * x

print("Standard function result:", square_standard(5))
print("Lambda function result:", square_lambda(5))

# Lambda with multiple arguments
add_lambda = lambda a, b: a + b
print("Lambda addition:", add_lambda(10, 15))
```

## 13. Functions: Practice Problems (Set 3)
1. Create a function to compute the factorial of a number `n`.
2. Create a function to return the largest of 3 numbers `a`, `b`, and `c`.

```python
# Solution 1: Factorial Function
def calculate_factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print("Factorial of 5 is:", calculate_factorial(5))

# Solution 2: Largest of 3 Numbers
def get_largest(a, b, c):
    if a > b and a > c:
        return a
    elif b > c:
        return b
    else:
        return c

largest_num = get_largest(3, 10, 5)
print("The largest number is:", largest_num)
```
