# Day 5: Object-Oriented Programming (OOP)
**Core Concepts:** Classes, Objects, Constructors, Attributes, Methods, and the 4 Pillars of OOP.

Object-Oriented Programming (OOP) models software as a collection of interacting "Objects"—just like real-world entities. It makes code modular, reusable, and easy to maintain by bundling data and behaviors together.

---

## 1. Classes & Objects
* **Class:** A blueprint or template. It describes what an object will look like, but it is *not* the object itself.
* **Object (Instance):** A concrete realization of a class. It contains actual data and occupies computer memory.

```python
# Classes & Objects

class Student:
    subject = "Python"
    college = "ABC"
    year = "4th year"

stu1 = Student()
stu2 = Student()

print(stu1.subject, stu1.college, stu1.year)
print(stu2.subject, stu2.college, stu2.year)
```

## 2. Constructors (`__init__`)
A constructor is a special method used to initialize newly created objects. Python calls it automatically whenever a new object is instantiated.
* **Default Constructor:** Takes no arguments other than `self`.
* **Parameterized Constructor:** Takes custom arguments to initialize unique values for each object.
* **`self`:** A special parameter that refers to the specific instance of the class that is calling the method.

```python
# Default Constructor
class Student:
    def __init__(self):
        print("constructor was called")

stu1 = Student()

# Parameterized Constructor
class StudentParameterized:
    def __init__(self, name):
        self.name = name

stu3 = StudentParameterized("Rahul")
stu4 = StudentParameterized("Harshita")

print(stu3.name, stu4.name)
```

## 3. Attributes & Methods

### Attributes (Variables)
* **Class Attributes:** Belong to the class itself. Shared by all objects (e.g., `college`).
* **Instance Attributes:** Belong uniquely to each object. Defined inside `__init__` (e.g., `name`, `gpa`).

### Methods (Functions)
* **Instance Methods:** Take `self` as the first argument. They access/modify the object's unique data.
* **Class Methods:** Use the `@classmethod` decorator and take `cls`. Used to work with class-level data.
* **Static Methods:** Use the `@staticmethod` decorator. They take neither `self` nor `cls` and behave like normal utility functions.

```python
# Attributes
class Student:
    college = "ABC college"        # class attribute
    def __init__(self, name, gpa): # instance attributes
        self.name = name
        self.gpa = gpa

stu1 = Student("Rahul", 8.7)
stu2 = Student("Harshita", 9.4)

print(f"{stu1.name} studying in {stu1.college} has {stu1.gpa} cgpa.")

# class attributes can also be accessed with class name
print(f"{stu2.name} studying in {Student.college} has {stu2.gpa} cgpa.") 

print("-" * 30)

# Methods
class Laptop:
    storage_type = "ssd"

    def __init__(self, RAM):
        self.RAM = RAM

    def get_info(self):   # Instance Method
        print(f"Laptop has {self.RAM} RAM & storage type is {self.storage_type}")

    @classmethod          # Class Method
    def get_storage_type(cls):
        print(f"storage type is {cls.storage_type}")

    @staticmethod         # Static Method
    def cal_discount(price, discount):
        final_price = price - (price * discount / 100)
        print(f"final price = {final_price}")

l1 = Laptop("16 GB")

l1.get_info() 
Laptop.get_storage_type()
l1.cal_discount(40_000, 10)
```

## OOP Pillar 1: Encapsulation
Encapsulation is bundling data and methods into a single class and **controlling access** to that data.
* **Public:** Accessible everywhere (`self.name`).
* **Private:** Indicated by a double underscore (`self.__balance`). Cannot be accessed directly from outside the class.

We use **Getters** and **Setters** to safely read or update private variables.

```python
# Encapsulation

class BankAccount:
    def __init__(self, name, balance):
        self.name = name            # public
        self.__balance = balance    # private

    def get_balance(self):         # getter
        return self.__balance

    def set_balance(self, new_balance):  # setter
        self.__balance = new_balance

acc1 = BankAccount("Rahul", 50000)
print("Initial Balance:", acc1.get_balance())
acc1.set_balance(60000)
print("Updated Balance:", acc1.get_balance())
```

## OOP Pillar 2: Inheritance
Inheritance allows a **Child** class to acquire properties and behaviors of a **Parent** class, enabling huge code reuse.
1. **Single Inheritance:** A child inherits from one parent.
2. **Multi-level Inheritance:** A child inherits from a parent, and another class inherits from that child.
3. **Multiple Inheritance:** A child inherits from more than one distinct parent class.

*Note: The `super()` keyword is used to call the parent class's constructor from within the child class.*

```python
# 1. Inheritance - Single Inheritance
class Employee:             # parent class
    start_time = "9AM"
    end_time = "5PM"

class Teacher(Employee):    # child class
    def __init__(self, subject):
        self.subject = subject

t1 = Teacher("Data Science")
print("Single:", t1.subject, t1.start_time, t1.end_time)

print("-" * 30)

# 2. Multi-level Inheritance
class EmployeeBase:             
    start_time = "9AM"
    end_time = "5PM"

class AdminStaff(EmployeeBase):     
    def __init__(self, role):
        self.role = role

class Accountant(AdminStaff):    
    def __init__(self, salary, role):
        super().__init__(role)
        self.salary = salary

acc1 = Accountant(50_000, "CA")
print("Multi-level:", acc1.salary, acc1.role, acc1.start_time, acc1.end_time)

print("-" * 30)

# 3. Multiple Inheritance
class TeacherBase:             
    def __init__(self, salary):
        self.salary = salary

class StudentBase():     
    def __init__(self, gpa):
        self.gpa = gpa

class TA(TeacherBase, StudentBase):    
    def __init__(self, name, salary, gpa):
        super().__init__(salary)       # Calls TeacherBase init
        StudentBase.__init__(self, gpa) # Explicitly calls StudentBase init
        self.name = name

ta = TA("Rahul", 50_000, 7.5)
print("Multiple:", ta.name, ta.salary, ta.gpa)
```

## OOP Pillar 3: Abstraction
Abstraction hides unnecessary implementation details and shows only the essential features to the user.
* **Abstract Class:** A blueprint that cannot be instantiated.
* **Abstract Method:** A method declared with `@abstractmethod` but left blank. The child classes *must* implement this method themselves.

```python
# Abstraction
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self): # Note: passing 'self' is required here for instance methods
        pass

class Lion(Animal):
    def make_sound(self):
        print("Roar!")

class Cow(Animal):
    def make_sound(self):
        print("Moo!")

lion = Lion()
lion.make_sound()

cow = Cow()
cow.make_sound()
```

## OOP Pillar 4: Polymorphism
Polymorphism is the ability of a single function, operator, or object to behave differently based on the context.
1. **Operator Overloading:** The same operator (`+`) behaves differently depending on the data type.
2. **Function Overriding:** A child class provides its own specific version of a method that already exists in the parent class.
3. **Duck Typing:** "If it looks like a duck and quacks like a duck, it must be a duck." Python doesn't care about the object's strict type, only that it has the required methods (like `speak()`).

```python
# Polymorphism

# Example - Operator Overloading
print("Operator Overload (Int):", 1 + 2)    # adds 2 numbers
print("Operator Overload (Str):", "1" + "2") # concatenates 2 strings

print("-" * 30)

# Function Overriding
class AnimalSound:
    def sound(self):
        print("Some generic sound")

class Dog(AnimalSound):
    def sound(self):
        print("Bark")

a = AnimalSound()
dog = Dog()

a.sound()    # Some generic sound
dog.sound()  # Bark

print("-" * 30)

# Duck Typing
class DuckDog:
    def speak(self):
        print("Bark")

class DuckCat:
    def speak(self):
        print("Meow")

class DuckRobot:
    def speak(self):
        print("Beep Boop")

def make_it_speak(entity):
    entity.speak()  # doesn’t care about type, just executes the method!

d = DuckDog()
c = DuckCat()
r = DuckRobot()

for e in [d, c, r]:
    make_it_speak(e)
```
