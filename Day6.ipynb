## 1. File I/O (Input / Output)
File I/O means reading data from external files and writing data to them. Python handles this natively using the `open()` function.

### Standard Syntax & Modes
`file = open("filename", "mode")`

**Common Modes:**
* `r` (Read): Opens for reading. Throws an error if the file doesn't exist. (This is the default).
* `w` (Write): Opens for writing. **Creates** a new file or **overwrites** an existing file completely.
* `a` (Append): Opens for writing, but adds content to the **end** of the file instead of overwriting it.
* `r+`, `w+`, `a+`: Combinations for reading and writing simultaneously.
* `b`: Binary mode (used for images, videos, etc., like `rb` or `wb`).

*CRITICAL RULE:* If you use `open()`, you must always use `file.close()` when you are done to free up system memory!

```python
# 1. Writing to a file (Creates the file if it doesn't exist)
f = open("demo_data.txt", "w")
f.write("Hello Machine Learning!\n")
f.write("This is Day 6.")
f.close() # Always close it!

# 2. Appending to a file
f = open("demo_data.txt", "a")
f.write("\nAppending a new line at the end.")
f.close()

# 3. Reading from a file
f = open("demo_data.txt", "r")
content = f.read() # Reads the entire file as a single string
print("--- File Content ---")
print(content)
f.close()
```

## 2. File I/O: The `with` Context Manager & OS Module
Manually writing `f.close()` every time is risky. If your program crashes before it reaches that line, the file remains locked in memory. 

**The Best Practice:** Use the `with open()` context manager. It automatically closes the file for you the moment the block of code finishes running, even if an error occurs inside it!

**Deleting Files:**
To delete a file, we import Python's built-in `os` module and use `os.remove("filename")`.

```python
import os

# Writing multiple lines using 'with open()'
lines_to_write = ["Line 1\n", "Line 2\n", "Line 3\n"]

with open("best_practice.txt", "w") as f:
    f.writelines(lines_to_write)
    # No f.close() needed! It closes automatically here.

# Reading specific lines
with open("best_practice.txt", "r") as f:
    print("First line:", f.readline().strip()) # .strip() removes the \n for clean printing
    print("Second line:", f.readline().strip())

# Reading all lines into a Python List
with open("best_practice.txt", "r") as f:
    all_lines = f.readlines()
    print("\nFile as a list:", all_lines)

# Deleting the files we just created to keep the directory clean
os.remove("demo_data.txt")
os.remove("best_practice.txt")
print("\nFiles deleted successfully using os.remove()")
```

## 3. Exception Handling
An exception is a runtime error. By default, if Python encounters an error (like dividing by zero, or trying to open a file that doesn't exist), the entire program crashes and stops working.

Exception handling allows us to catch these errors and manage them gracefully so the program keeps running.

* `try`: Put the risky code here that might cause an error.
* `except`: If an error happens in the `try` block, the program jumps here instead of crashing.
* `else`: Executes ONLY if no error occurred in the `try` block.
* `finally`: Executes absolutely always (error or no error). Perfect for cleanup tasks.

```python
print("--- Basic Try/Except ---")
try:
    # Risky code
    result = 10 / 0
except:
    print("An error occurred! But the program didn't crash.")

print("\n--- Catching Specific Exceptions ---")
try:
    with open("missing_file.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Error: That file does not exist!")
except ZeroDivisionError:
    print("Error: You tried to divide by zero!")

print("\n--- The Else and Finally Blocks ---")
try:
    val = 10 / 2
except ZeroDivisionError:
    print("Math Error!")
else:
    print(f"Success! The result is {val}")
finally:
    print("Execution completed. Cleaning up resources...")
```

## 4. List Comprehensions
List Comprehension is a short, elegant way to create lists in Python. It replaces long, multi-line `for` loops with a single, highly readable line of code.

### Syntax Variations:
1. **Basic:** `[expression for item in iterable]`
2. **With Condition (Filtering):** `[expression for item in iterable if condition]`
3. **With If-Else (Modification):** `[expression_if_true if condition else expression_if_false for item in iterable]`

```python
# Traditional way to create a list of numbers 1 to 5
old_list = []
for x in range(1, 6):
    old_list.append(x)

# 1. Basic List Comprehension
nums = [x for x in range(1, 6)]
print("Basic Comprehension:", nums)

# 2. List Comprehension with an IF condition (Get only even numbers)
# Logic: Give me 'x', for every 'x' in 1-10, ONLY IF x is divisible by 2.
evens = [x for x in range(1, 11) if x % 2 == 0]
print("Filtered Evens:", evens)

# 3. List Comprehension with IF-ELSE
# Logic: Give me "Even" if divisible by 2, ELSE give me "Odd", for every 'x' in 1-5.
labels = ["Even" if x % 2 == 0 else "Odd" for x in range(1, 6)]
print("If-Else Labels:", labels)
```

## 5. Working with the JSON Module
JSON (JavaScript Object Notation) is a lightweight data format used globally to exchange data between programs, APIs, and websites. It looks almost exactly like a Python Dictionary.

Python has a built-in `json` module to handle this.
* **`dumps()`:** Converts a Python Dictionary into a JSON String.
* **`loads()`:** Converts a JSON String back into a Python Dictionary.
* **`dump()`:** Writes a Python Dictionary directly into a `.json` file.
* **`load()`:** Reads a `.json` file directly into a Python Dictionary.

```python
import json

print("--- 1. Python to JSON String ---")
python_dict = {
    "name": "Anant",
    "role": "AI/ML Engineer",
    "skills": ["Python", "Data Science"]
}
# Convert to JSON string
json_string = json.dumps(python_dict)
print("JSON String Type:", type(json_string))
print(json_string)

print("\n--- 2. JSON String to Python ---")
# Pretend we received this string from an API
received_json = '{"name": "John", "age": 25}' 
parsed_dict = json.loads(received_json)
print("Parsed Object Type:", type(parsed_dict))
print("Accessed Name:", parsed_dict["name"])

print("\n--- 3. Writing to and Reading from a JSON File ---")
# Writing to a file (using indent=4 makes the file highly readable)
with open("profile.json", "w") as f:
    json.dump(python_dict, f, indent=4)
print("Successfully wrote data to profile.json")

# Reading from the file
with open("profile.json", "r") as f:
    loaded_data = json.load(f)
    print("Successfully read data from profile.json:")
    print(loaded_data)

# Cleanup
os.remove("profile.json")
```
